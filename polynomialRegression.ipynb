```python
import pandas as pd
import numpy as np
train_data=pd.read_csv(r"polynomial_regression_train.csv")
test_data=pd.read_csv(r"polynomial_regression_test.csv")
x_test = test_data.drop(['ID'],axis=1).values
x_DataFrame=train_data.iloc[:,:-1].drop(['ID'],axis=1)
x=x_DataFrame.values
y_DataFrame=train_data.iloc[:,-1]
y=y_DataFrame.values
y_mean=np.mean(y)
def standardize(X):
    return (X - np.mean(X, axis=0)) / np.std(X, axis=0)

x = standardize(x)
x_test = standardize(x_test)
rows,columns=x_DataFrame.shape
print(rows)
print(columns)
train_data.head()
```

```python
def generate_polynomial_features(X, degree):
    n_samples, n_features = X.shape
    poly_features = np.ones((n_samples, 1))
    for d in range(1, degree + 1):
        for i in range(n_features):
            poly_features = np.hstack((poly_features, (X[:, i:i+1] ** d)))
            for j in range(i + 1, n_features):
                new_feature = (X[:, i:i+1] ** (d - 1)) * (X[:, j:j+1])
                poly_features = np.hstack((poly_features, new_feature))
    return poly_features
degree = 3
x_poly = generate_polynomial_features(x, degree)
x_T = x_poly.T

print(x_poly.shape)
```

```python
coefficients = np.linalg.pinv(x_T @ x_poly) @ x_T @ y
coefficients=coefficients.reshape(x_poly.shape[1],1)
print(coefficients.shape)
```

```python

x_test_poly = generate_polynomial_features(x, degree)
# Fit the model using normal equation
x_test_T = x_test_poly.T
```

```python
predicted_target=x_test_poly@coefficients
```

```python
predicted_target_df=pd.DataFrame(predicted_target,columns=['target'])
predicted_target_df_=test_data.join(predicted_target_df)
predicted_target_df_.to_csv('predicted_target_polynomial_regression.csv',index=False)
```
