```python
import pandas as pd
import numpy as np
unsupervised_data=pd.read_csv(r"unsupervised_data.csv")
unsupervised_data.head()
data=unsupervised_data.drop('ID',axis=1).values
print(data.shape,data.shape[0])
```

```python
k=6
np.random.seed(42)
indices = np.random.choice(len(data), k, replace=False)
centroids=data[indices]
epochs=1000
for i in range(epochs):
    old_centroids=centroids
    min_index=[]
    for row in data:
        list=[]
        for centroid in centroids:
            dist = np.sqrt(np.sum((row - centroid) ** 2))
            list.append(dist)
        data_=np.array(list)
        min_index_=np.argmin(data_)
        list_of_index=min_index_.tolist()
        min_index.append(list_of_index)
    print(min_index)
    cluster=np.array(min_index)
    cluster_df=pd.DataFrame(cluster)
    cluster_df.head
    updated_centroid=np.array([data[cluster == n].mean(axis=0) for n in range(k)])
    if np.all(centroids == old_centroids):
        break
```

```python
predicted_cluster_through_k_mean_clustering_=unsupervised_data.join(cluster_df)
# predicted_cluster_through_k_mean_clustering_.to_csv('predicted_cluster_through_k_mean_clustering_.csv')
predicted_cluster_through_k_mean_clustering_.head()
```
