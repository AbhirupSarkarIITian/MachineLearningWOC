```python
import numpy as np
import pandas as pd
# Normalize the features
def normalize(X):
    return (X - X.mean()) / X.std()
```

```python

def one_hot_encode(y, num_classes):
    return np.eye(num_classes)[y]
def softmax(logits):
    exp_logits = np.exp(logits - np.max(logits, axis=1, keepdims=True))  # Stability trick
    return exp_logits / np.sum(exp_logits, axis=1, keepdims=True)
def cross_entropy_loss(y_true, y_pred):
    m = y_true.shape[0]
    return -np.sum(y_true * np.log(y_pred + 1e-9)) / m
```

```python
def train_softmax_regression(X, y, num_classes, learning_rate, epochs):
    m, n = X.shape
    W = np.zeros((n, num_classes))
    b = np.zeros(num_classes)
    for epoch in range(epochs):
        logits = X @ W + b
        y_pred = softmax(logits)
        loss = cross_entropy_loss(y, y_pred)
        dW = (1 / m) * (X.T @ (y_pred - y))
        db = (1 / m) * np.sum(y_pred - y, axis=0)
        W -= learning_rate * dW
        b -= learning_rate * db
        if epoch % 100 == 0:
            print(f"Epoch {epoch}, Loss: {loss}")
    return W, b
```

```python

def predict(X, W, b):
    logits = X @ W + b
    y_pred = softmax(logits)
    return np.argmax(y_pred, axis=1)
def accuracy(y_true, y_pred):
    return np.mean(y_true == y_pred) * 100
```

```python

import pandas as pd
import numpy as np
train_data=pd.read_csv(r"multi_classification_train.csv")
test_data=pd.read_csv(r"multi_classification_test.csv")
test_data.head()
x=train_data.iloc[:,:-1]
X_train=train_data.drop('ID',axis=1).values
print(X_train.shape)
y_train=train_data.iloc[:,-1].values
X_test=test_data.values
X_train = normalize(X_train)
X_test = normalize(X_test)
```

```python
num_classes = len(np.unique(y_train))
y_train_one_hot = one_hot_encode(y_train, num_classes)
learning_rate = 0.01
epochs = 5000
W, b = train_softmax_regression(X_train, y_train_one_hot, num_classes, learning_rate, epochs)
y_pred = predict(X_train, W, b)
```
