```python
import pandas as pd
import numpy as np
train_data=pd.read_csv(r"binary_classification_train.csv")
test_data=pd.read_csv(r"binary_classification_test.csv")
train_data.shape
```

```python
x_DataFrame=train_data.iloc[:,:-1]#all the features are extracted
x_train=x_DataFrame.drop(['ID'],axis=1).values#Converting the datframe in array to perform numpy operations
y_DataFrame=train_data.iloc[:,-1]#including target value
y_train=y_DataFrame.values
x_test=test_data.drop(['ID'],axis=1).values
print(y_train.ndim)
print(x_train.ndim)
print(x_test.ndim)
y_train=y_train.reshape(48000,1)
print(y_train.ndim)
print(x_train.shape)
```

```python
def sigma(z):
    z = np.clip(z, -500, 500)
    return 1 / (1 + np.exp(-z))
learning_rate=0.1
epochs=1000
np.random.seed(50)
w = np.random.randn(20, 1)
b=np.zeros((1,))
z=x_train@w + b
print(z.shape)
epsilon = 1e-15  # A very small value
z = np.clip(z, epsilon, 1 - epsilon)
loss=-np.mean(y_train * np.log(z) + (1 - y_train) * np.log(1 - z))
print(loss)
```

```python
for i in range(epochs):
    #keep w updating
    gradient=((x_train.T)@(sigma(x_train@w)-y_train))/48000
    w=w-learning_rate*gradient
    b-=learning_rate*np.sum(sigma(z) - y_train) / 48000
    z=x_train@w+b
    print(i)
```

```python
epsilon = 1e-15  # A very small value
z = np.clip(z, epsilon, 1 - epsilon)
losses=-np.mean(y_train * np.log(z) + (1 - y_train) * np.log(1 - z))
print(losses)
```

```python
def prediction(probabilities):
   threshold=np.full((12000, 1), 0.5)
   return (probabilities >= threshold).astype(int)
```

```python
z=x_test@w+b
predicted_class=prediction(sigma(z))
```

```python
predicted_class_=pd.DataFrame(predicted_class)
predicted=test_data.join(predicted_class_).to_csv("predicted_class_Binary_classification.csv",index=False)
```
